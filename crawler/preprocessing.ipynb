```python
# load json file
import json, pandas as pd
 
with open('author_list.json') as f:
    js = json.loads(f.read())
author_df = pd.DataFrame(js)

with open('processed_paper_dict.json') as f:
    js = json.loads(f.read())
paper_df = pd.DataFrame(js).T
```

```python
# combining data collected from google scholar to the columns

paper_df = paper_df.rename(columns = {"google_schorlar_metadata": "google_scholar_metadata"})
paper_df['year'] = paper_df['google_scholar_metadata'].apply(lambda x: x['게시 날짜'][:4] if '게시 날짜' in x else None)
paper_df['abstract'] = paper_df['google_scholar_metadata'].apply(lambda x: x['설명'] if '설명' in x else None)
paper_df['authors'] = paper_df['google_scholar_metadata'].apply(lambda x: x['저자'] if '저자' in x else None)
paper_df['referenced_list'] = paper_df['google_scholar_metadata'].apply(lambda x: x['전체 인용횟수'].split()[0][:-1] if '전체 인용횟수' in x else None)
paper_df['conference'] = paper_df['google_scholar_metadata'].apply(lambda x: x['컨퍼런스'].split(',')[0].split(':')[0] if '컨퍼런스' in x else None)
paper_df['journal'] = paper_df['google_scholar_metadata'].apply(lambda x: x['저널'] if '저널' in x else None)
paper_df['issn'] = paper_df['crossref_json'].apply(lambda x: eval(str(x['ISSN']))[0] if 'ISSN' in x else None)

paper_df
```

```python
# adding the keywords to the columns of the papers

paper_df['keywords'] = ''

for index, row in paper_df.iterrows():
    if pd.notna(row['authors']):
        authors = row['authors'].split(', ')
        expertise_lists = []
        # find the expertise list of each others add to keywords
        for author in authors:
            if author in author_df['name'].values:
                expertise = author_df.loc[author_df['name'] == author, 'expertise_list'].iloc[0]
                if isinstance(expertise, list):
                    expertise_lists.extend([item.lower() for item in expertise])
        if expertise_lists:
            paper_df.at[index, 'keywords'] = ', '.join(experience for experience in set(expertise_lists))

paper_df

```

```python
# extracting information for the author table by extracting the author type

author_paper_df = pd.DataFrame(columns=['DOI', 'author_name', 'author_type'])

for index, row in paper_df.iterrows():
    doi = row['DOI']
    authors = row['authors']
    
    # check for missing values
    if pd.notna(authors):
        authors = authors.split(', ')
        for i, author in enumerate(authors): # find the the author type
            author_type = 'first author' if i == 0 else 'corresponding author' if i == len(authors) - 1 else 'author'
            author_paper_df = pd.concat([author_paper_df, pd.DataFrame({'DOI': [doi], 'author_name': [author.strip()], 'author_type': [author_type]})],
                                        ignore_index=True)


author_paper_df = author_paper_df.dropna(subset=['DOI'])
author_paper_df
```

```python
# getting rid of the duplicate rows
paper_df = paper_df.drop_duplicates(subset=['DOI'])
author_paper_df = author_paper_df.drop_duplicates(subset=['DOI', 'author_name', 'author_type'])
```

```python
# getting rid of the duplicate rows
author_df = author_df.drop_duplicates(subset=['name'])
```

```python
# importing as json file
paper_df.to_json('paper_final.json',orient="index", indent=2)
author_paper_df.to_json('author_paper.json',orient="index", indent=2)
```

```python
# loading json file for conferences
with open('journal_conference_dict.json') as f:
    js = json.loads(f.read())
conf_df = pd.DataFrame(js).T

conf_df
```

```python
# preprocessing the strings with no blanks and adjust it into formats

conf_df['ISSN'] = conf_df['ISSN'].apply(lambda x: x.strip())
conf_df['url_list'] = conf_df['url_list'].apply(lambda x: x[0] if len(x)!=0 else None)

conf_df.drop_duplicates(subset=['ISSN'])

# only extract information needed
conf_df = conf_df[['name', 'ISSN', 'url_list', 'Country']]

conf_df
```

```python
# saving it to json
conf_df.to_json('conference.json',orient="index", indent=2)
```
